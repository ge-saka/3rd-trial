```python
nltk.download('punkt') # a tokenizer that is used for dividing text into words or sentences
nltk.download('stopwords') # to filter out common words that are considered less meaningful for text processing
nltk.download('wordnet') # group words that share common meaning
```

```python
## Lets define the stop words and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
# Inspect the prompt  column to check the overview of the words 
print(df['prompt'].sample(10))
```

From the inspection of the 'Prompt' column, we need to clean the data by removing spaces, digits in the text,
removing irrelevant punctuations, removing multiple spaces and removing trailing spaces.<br>
We will also use the stop words and lematizer for two purposes namely:<br>
1. Filtering out common words that do not add significant value to the analysis of text.<br>
2. Reducing words to their base form to ensure that different forms of a word are treated as the same token.<br>
This is important inorder to reduce skewness by removing common words that do not affect the sentiment.It also helps in classifying text more effectively by focusing on the significant terms.

```python
# Define a function for basic text cleaning
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^a-zA-Z\s]', '', text, re.I|re.A)  # Remove special characters
    text = text.lower()  # Convert to lowercase
    text = text.strip()  # Remove whitespaces
    # Tokenize
    tokens = word_tokenize(text)
    # Removing stop words and lemmatizing
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python
df['cleaned_prompt'] = df['prompt'].apply(clean_text)
df['cleaned_generation'] = df['generation'].apply(clean_text)
df.head()
```

<span style='color:Green'>***VADER(Valence Aware Dictionary and Sentiment Reasoner)***</span>

The VADER  sentiment analysis tool is a rule-based sentiment analysis tool specifically and mostly  designed for social media texts.<br>
It can analyze the sentiment of text as positive, neutral, or negative, and also provides a sentiment polarity score.

```python
#!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
# Initialize VADER SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

'Polarity_scores(prompt): Returns a dictionary with keys 'neg', 'neu', 'pos', and 'compound' representing the negative, neutral, positive, and overall sentiment scores respectively.

```python
# Apply VADER sentiment analysis to the 'prompt' column
df['prompt_sentiment_scores'] = df['cleaned_prompt'].apply(lambda text: analyzer.polarity_scores(text))
```

```python
# Apply VADER sentiment analysis to the 'generation' column
df['generation_sentiment_scores'] = df['cleaned_generation'].apply(lambda text: analyzer.polarity_scores(text))
```

***Sentiment Threshold***

The threshold for 0.05 is a common practice in sentiment analysis. However for us, we are going to adjust our threshold to 1/3 of the score range i.e., 1- to 1. We are going to use that threshold to classify the compound scores for promt and generation respectively.

```python
# Define the thresholds
neg_threshold = -1/3  # -0.333...
pos_threshold = 1/3   # 0.333...

# Categorize sentiments for prompted texts
def categorize_sentiment(score):
    if score >= pos_threshold:
        return 'Positive'
    elif score <= neg_threshold:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
## APPLYING THE FUNCTION TO THE PROMPT TEXT P.COMPOUND SCORE
df['prompt_sentiment_category'] = df['p.compound'].apply(categorize_sentiment)
```

```python
## APPLYING THE FUNCTION TO THE GENERATION TEXT G.COMPOUND SCORE
df['generation_sentiment_category'] = df['g.compound'].apply(categorize_sentiment)
```

<span style='color:Blue'>***Latent Dirichlet Allocation(LDA)***</span>

LDA(Latent Dirichlet Allocation) is a method used to uncover the underlying topics in a collection of text documents.<BR>In out case,based on the collection of texts we have in the 'cleaned_prompt' column, LDA is able to categorize them into 3 categories of positive,negative,neutral.<BR>
In Technical terms;<BR>
1. LDA analyzes a collection of documents to identify clusters of words that frequently appear together. These clusters represent topics.<BR>
2. Assigns Topics: Each document is then assigned a distribution over these topics. <BR>
3. Displays Words per Topic: For each topic, LDA provides a list of words that are most associated with that topic.

```python
from gensim import corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
from gensim import matutils
from gensim.models import TfidfModel
from nltk.tokenize import word_tokenize
import re

# Step one:Convert texts to a list of strings
texts_as_strings = df['cleaned_prompt'].tolist()
```

```python
# We then Vectorize the strings using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(texts_as_strings)
```

```python
# Convert to gensim format;this Creates a Gensim dictionary from this corpus
corpus_tfidf = matutils.Sparse2Corpus(tfidf_matrix, documents_columns=False)
dictionary_tfidf = corpora.Dictionary.from_corpus(corpus_tfidf)
```

```python
# Define and train the LDA model
lda_model_tfidf = LdaModel(corpus_tfidf, num_topics=3, id2word=dictionary_tfidf, passes=5)
```

```python
# Print a snippet of the topics
for idx, topic in lda_model_tfidf.print_topics(-1):
    print(f'Topic: {idx} \nWords: {topic}')
```

```python
# Inspect topics the  3 topics 
for topic_id in range(lda_model_tfidf.num_topics):
    topic_words = lda_model_tfidf.show_topic(topic_id, topn=10)  # Get the top 10 words for each topic
    print(f"Topic {topic_id}:")
    for word, prob in topic_words:
        print(f"{word}: {prob:.4f}")
    print("\n")
```
